```python
import numpy as np
import pandas as pd


goog = pd.read_csv("goog.csv")
spy = pd.read_csv("SPY.csv")


df = pd.DataFrame({
    "spy": spy["Return"],
    "goog": goog["Return"]
})

df = df.replace([np.inf, -np.inf], np.nan).dropna()

return_spy = df["spy"].to_numpy()
return_goog = df["goog"].to_numpy()

m, b = np.polyfit(return_spy, return_goog, deg=1)

print(m)



mean_spy = np.mean(return_spy)
print("SPY mean:", mean_spy)

sd_spy = np.std(return_spy)

mean_goog = np.mean(return_goog)
sd_goog = np.std(return_goog)
print("GOOG mean:", mean_goog)

pos_return_spy = return_spy[return_spy>0]
neg_return_spy = return_spy[return_spy<0]
print("SPY: postive returns:", len(pos_return_spy), "negative returns:", len(neg_return_spy))

pos_return_goog = return_goog[return_goog>0]
neg_return_goog = return_goog[return_goog<0]
print("GOOG: postive returns:", len(pos_return_goog), "negative returns:", len(neg_return_goog))

outliers_spy = return_spy[abs(mean_spy-return_spy)/sd_spy >2]
print("SPY outliers", outliers_spy, "length:", len(outliers_spy))

outliers_goog = return_goog[abs(mean_goog-return_goog)/sd_goog >2]
print("GOOG outliers", outliers_goog, "length:", len(outliers_goog))
```
